```python
from ultralytics import YOLO
from weighted_trainer import WeightedDetectionModel
```

```python
model = YOLO("runs/detect/runs/Front/front_yolo26s/weights/best.pt")
source = "I:\ProgramData\课堂专注度项目\P1_YOLO_Pose_Attention\学生全景_10.30.04-p01-80.mp4"
```

```python
import cv2

# 用 stream=True，结果逐帧处理后立即释放，不积累内存
results = model.predict(
    source,
    stream=True,    # ← 关键，必须加
    save=False,
    show=False,     # Jupyter 里 show=True 会有问题，改用 cv2 显示
    conf=0.2,       # 降低置信度阈值，让少数类也能显示
)

for result in results:
    frame = result.orig_img                    # 原始帧
    annotated = result.plot()                  # 画上检测框

    cv2.imshow("Detection", annotated)
    if cv2.waitKey(1) & 0xFF == ord("q"):      # 按 q 退出
        break

cv2.destroyAllWindows()
```
